Class Activation Maps with ResNet18

1. Introduction
The aim of this lab was to investigate interpretability in convolutional neural networks using Class Activation Maps (CAM). By applying SmoothGradCAM++ to a pretrained ResNet18 model, we visualized which image regions contributed most to the model’s predictions. Positive and negative examples were used for three classes — goldfish, labrador, and ambulance — to evaluate both correct classifications and misclassifications.

2. Methodology
- Dataset: Images were organized into subfolders under `lab2/data` for each class. Each folder contained positive examples (true class images) and negative examples (similar but incorrect objects).  
- Model: ResNet18 pretrained on ImageNet was used without fine-tuning.  
- Preprocessing: All images were resized to 224×224, converted to tensors, and normalized with ImageNet mean and standard deviation.  
- CAM technique**: SmoothGradCAM++ was applied to highlight discriminative regions in the input images.  
- Implementation**: Predictions and CAM overlays were generated for each image and saved into `lab2/results`.

3. Results

Goldfish
- Positive example (goldfish.png) → Correctly predicted as goldfish. CAM overlay (`goldfish_goldFish_cam.png`) highlighted the fish’s body, showing the model focused on relevant features.  
- Negative example (salmonFish.png) → Misclassified as puffer. CAM overlay (`goldfish_salmonFish_cam.png`) showed attention on the fish’s shape, but the model confused it with another aquatic species.

Labrador
- Positive example (dog.png) → Predicted as Pembroke (dog breed). CAM overlay (`labrador_dog_cam.png`) highlighted the dog’s head and torso, indicating recognition of canine features but mislabelling the breed.  
- Negative example (germanshepered.png) → Predicted as German shepherd. CAM overlay (`labrador_germanshepered_cam.png`) showed focus on fur texture and body shape, revealing difficulty distinguishing between similar breeds.

Ambulance
- Positive example (ambulance.png) → Correctly predicted as ambulance. CAM overlay (`ambulance_ambulance_cam.png`) highlighted the vehicle outline and windows.  
- Negative example (car.png) → Misclassified as *crash helmet. CAM overlay (`ambulance_car_cam.png`) focused on wheels and body shape, suggesting reliance on superficial similarities and producing an unexpected misclassification.

4. Analysis
- Positive examples**: CAM overlays aligned with the target objects, confirming the model’s ability to identify discriminative regions.  
- Negative examples: Misclassifications revealed reliance on misleading features — fish shape, dog fur, or vehicle outlines — leading to confusion.  
- Interpretability: CAM provided valuable insight into why predictions were correct or incorrect. For example, the “salmonFish” misclassification showed the model’s bias toward shape rather than species-specific details.  
- Limitations: The pretrained ResNet18 was not fine-tuned for these specific classes, which explains breed-level confusion (Labrador vs. German shepherd) and unusual misclassifications (car → crash helmet).  

5. Critical Reflection
CAM proved essential for critically evaluating the model’s strengths and weaknesses. Without visualization, misclassifications such as “car → crash helmet” would appear arbitrary. With CAM, it became clear that the model was focusing on wheels and outlines, which explains the error. This highlights both the value of interpretability tools and the limitations of pretrained models when applied to custom datasets. For VG-level analysis, it is important to show not only what the model predicted, but why it made those predictions — CAM enabled this deeper understanding.

6. Conclusion
This experiment demonstrated that CAM is a powerful tool for interpreting deep learning predictions. Positive examples showed the model’s focus on relevant features, while negative examples highlighted its reliance on misleading similarities. To improve performance, future work could involve fine-tuning the model on a custom dataset, expanding the dataset with more diverse examples, and comparing different CAM methods (GradCAM, LayerCAM, etc.).
