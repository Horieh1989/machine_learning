
**Method**
In this project, I built a content-based movie recommendation system using data from `movies.csv`, `ratings.csv`, and `tags.csv`. The goal was to recommend five movies given one input movie. Based on the assignment instructions, I chose to combine metadata from `movies.csv` with user-generated tags from `tags.csv`, and to use summary information from `ratings.csv` to strengthen the dataset.

The assignment notes that recommendation can be based either on rating prediction or on movie similarity. It also points out that better results can be achieved by combining the metadata in `tags.csv` with the genres in `movies.csv` and using TF-IDF vectorization. I followed that direction and used a content-based filtering approach.

**Data Preparation**
I first explored the three main files:
- `movies.csv` contains `movieId`, `title`, and `genres`
- `ratings.csv` contains user ratings and is very large, with more than 33 million rows
- `tags.csv` contains user-generated text tags that describe movies in a more detailed way than genres alone

From the exploration, I observed that:
- `genres` are useful but quite broad
- `ratings.csv` has very high coverage, but is too large to use directly in a full user-movie matrix on a normal computer
- `tags.csv` contains rich text information and is suitable for basic natural language processing
- tags are available for many movies, but not all of them, so it was useful to combine tags with genres rather than rely on tags alone

To build a common dataset, I created a movie-level table with one row per movie. I cleaned the tag text by:
- removing missing tags
- converting all tags to lowercase
- removing empty strings
- combining all tags for the same movie into one text field

I also summarized the ratings per movie by calculating:
- `rating_count`
- `avg_rating`

After that, I merged the information into one dataset containing:
- `movieId`
- `title`
- cleaned genres
- aggregated tag text
- number of ratings
- average rating

**Feature Construction**
To represent each movie as text, I transformed the `genres` column into plain text by replacing the `|` separator with spaces and converting everything to lowercase. I then combined the cleaned genres and aggregated tags into one text feature called `text_features`.

This was done because genres provide broad category information, while tags provide more detailed semantic information such as tone, themes, style, and plot elements. Combining them gives a richer representation of each movie.

**Natural Language Processing Technique**
As suggested in the assignment, I used TF-IDF vectorization on the combined text features. TF-IDF stands for Term Frequency-Inverse Document Frequency. This technique converts text into numerical vectors by giving higher weight to words that are important for a specific movie and lower weight to words that appear very frequently across many movies.

This made it possible to compare movies mathematically based on the words in their genres and tags.

**Recommendation Method**
To generate recommendations, I used cosine similarity on the TF-IDF vectors. Cosine similarity measures how similar two movie vectors are in direction. If two movies have similar genres and tags, their vectors will be close and their cosine similarity will be high.

Given an input movie title, the system:
- finds that movie in the dataset
- computes similarity between that movie and all others
- ranks the movies by similarity
- returns the five most similar movies

To avoid unnecessary memory use, I did not compute the full movie-to-movie similarity matrix for all movies at once. Instead, I computed similarity only between the selected input movie and the rest of the dataset. This was important because the dataset contains more than 86,000 movies.

**Example Result**
For the input movie `Toy Story (1995)`, the system recommended:
- `Toy Story 2 (1999)`
- `A Bug’s Life (1998)`
- `Toy Story 3 (2010)`
- `Finding Dory (2016)`
- `Monsters, Inc. (2001)`

These recommendations are reasonable because they share similar genres and related tag-based descriptions, such as animation, family-oriented themes, and similar storytelling style.

**Why This Method Was Chosen**
This method matches the assignment well because:
- it uses the cross-referenced movie data from multiple files
- it applies a natural language processing technique to `tags.csv`
- it improves over genre-only similarity by including richer textual metadata
- it avoids the heavy computational cost of full collaborative filtering on the entire ratings dataset

**Limitations**
This method is still limited in some ways:
- movies without tags rely mostly on genres, which are less informative
- it does not use full user-rating patterns in the way collaborative filtering does
- recommendations are based on content similarity, not personalized user taste
